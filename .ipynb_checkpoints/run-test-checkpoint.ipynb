### 将所有细胞按照样本分门别类的放在不同的列表中

```python
def getSampleName(pic_path):
    """
    function：
        提取一个图片路径pic_path的样本名称
    parameters：
        pic_path：图片全路径
    """
    pic_name = pic_path.replace(os.path.dirname(pic_path)+"/","")
    return pic_name.split('_')[0]
    
# 将所有图片根据样本名单独放在一个列表中
def splitSampleByName(X_train):
    """
    function：
        将输入列表中的细胞图片按照样本分门别类的放在不同的列表中
    parameters：
        X_train：细胞图片路径列表
    """
    sample_names = []
    result = []
    for i in range(len(X_train)):
        sample = X_train[i]
        sample_name = getSampleName(sample)
        if sample_name not in sample_names:
            sample_names.append(sample_name)  # 插入样本名称 
            result.append([sample])  # 插入图片全路径
        else:
            index_ = sample_names.index(sample_name)
            result[index_].append(sample)
    assert len(sample_names)==len(result), "函数splitSampleByName中样本名称长度不相等！"
    return sample_names,result

# 将所有图片根据样本名单独放在一个列表中
def splitSampleByName_(X_train):
    """
    function：
        将输入列表中的细胞图片按照样本分门别类的放在不同的列表中
    parameters：
        X_train：细胞图片路径列表
    """
    return ["全部数据集"],[X_train]

sampleNames,samplesSplitByName = splitSampleByName(X_train)
samplesSplitByName_len = [len(samplesSplitByName[i]) for i in range(len(samplesSplitByName))]  # 每个样本所包含的细胞数量
# print (samplesSplitByName_len)
```

```python
import PIL.Image as Image
import os
import matplotlib.pyplot as plt

def splice(image_list_,is_training=True): 
    """
    function：
        对图像进行拼接，将若干小图拼接为一个大图
    parameters：
        image_list_：若干小图路径列表
        is_training：标识是否是训练状态
    """
    image_size = cfg0.image_size
    image_row = cfg0.image_row
    image_column = cfg0.image_column
    read_time = 0
    
    image_list = image_list_[:]  # 列表副本
    # assert len(image_list)==image_column*image_row, "splice函数中图片总数量不对应！" 
    result = Image.new('RGB', (image_column * image_size, image_row * image_size)) #创建一个新图
    k = 0
    for i in range(image_row):
        for j in range(image_column):
            if k>=len(image_list):
                img = Image.new('RGB', (image_column * image_size, image_row * image_size),color=0)
            else:
                # img = Image.open(image_list[k])
                time_start = time.time()
                if is_training: 
                    img = 255*dataenhance.loadAndEnhanceTrainData(image_list_[k])  # 数据增强，并逆归一化  
                    img = Image.fromarray(img.astype('uint8')).convert('RGB')  
                else:
                    img = 255*dataenhance.loadAndEnhanceTestData(image_list_[k])  # 数据增强，并逆归一化 
                    img = Image.fromarray(img.astype('uint8')).convert('RGB')  
                time_end = time.time()
                read_time = read_time + (time_end-time_start)
            result.paste(img,(i*image_size,j*image_size))
            k = k+1
    # result.save(pic_path)
    # plt.imshow(result)
    # plt.show()
    return np.array(result),read_time

index = 0
image_list = utils.utils().myShuffle(samplesSplitByName[index])[0:100] # 图片集地址
img_tmp,_ = splice(image_list,is_training=True)
```

### 单个训练图片采样函数

```python
import random

def random_index(weight):
    """
    function：
        根据权值列表为概率返回列表的索引下标
        参考至：https://blog.csdn.net/qq_30008595/article/details/88674893 
    parameters：
        weight：权值列表
    """
    start = 0
    index = 0
    randnum = random.randint(1, sum(weight))
    for index, scope in enumerate(weight):
        start += scope
        if randnum <= start:
            break
    return index

weight = [90,10]
xxx = [random_index(weight) for i in range(10)]
print(sum(xxx))
print(xxx)
```

```python
from itertools import chain

def getSingleTrainPic(samplesSplitByName,is_training=True):
    """
    function：
        获取单个训练图片：随机的对训练集中的样本进行采样（row*column个），然后将采样图拼接为1280×1280大图
    paratmeters：
        samplesSplitByName：str类型二维列表，根据样本名称对训练集分门别类放在不同列表
        is_training：标识是否是训练状态
    """
    # 首先将所有样本的细胞随机打乱
    # for i in range(len(samplesSplitByName)):
    #     samplesSplitByName[i] = utils_.myShuffle(samplesSplitByName[i])

    # sampleNum = len(samplesSplitByName)  # 样本总数
    # sampleIndex = random.randint(0, sampleNum)  # 随机选择一个样本
    weight = [len(samplesSplitByName[i]) for i in range(len(samplesSplitByName))]  # 各个样本包含的细胞数量
    sampleIndex = random_index(weight)  # 以细胞数量为概率随机生成一个样本
    picNumOfSample = len(samplesSplitByName[sampleIndex])  # 所选择的样本中细胞数量
    if picNumOfSample<cfg0.image_row*cfg0.image_column:  # 表示不够拼接成一个完整的图像
        XX = samplesSplitByName[sampleIndex][:]
        yy = utils_.geneLabel(XX,labels)
        yy = list(chain.from_iterable(yy))  # 将二维列表合并为一维列表
        
        # 计算补充的黑体图像的个数（看做Junk类别）
        blackJunkNum = cfg0.image_row*cfg0.image_column - picNumOfSample
        yy = yy + blackJunkNum*[1 if i==cfg0.junk_index else 0 for i in range(len(labels))]  # 标签中加入拼接图像中补充的黑色背景图像
        result,_ = splice(XX)
    else:
        # 在[0,picNumOfSample-cfg0.image_row*cfg0.image_column]之间生成随机数，作为切片的起点
        start = random.randint(0,picNumOfSample-cfg0.image_row*cfg0.image_column)
        XX = samplesSplitByName[sampleIndex][start:start+cfg0.image_row*cfg0.image_column]
        yy = utils_.geneLabel(XX,labels)
        yy = list(chain.from_iterable(yy))  # 将二维列表合并为一维列表
        result,_ = splice(XX,is_training)
    return result,yy  # 返回训练/测试样本，及标签
```

### 获取一个batch的样本

```python
def getMiniBatch(samplesSplitByName,is_training=True):
    """
    function：
        返回一个经过数据增强和拼接后的训练/预测Batch
    paratmeters：
        samplesSplitByName：str类型二维列表，根据样本名称对训练集分门别类放在不同列表
        is_training：标识是否是训练状态
    """
    XXs = []
    yys = []
    for i in range(cfg0.batch_size):
        XX,yy = getSingleTrainPic(samplesSplitByName,is_training) 
        XXs.append(np.array(XX))
        yys.append(yy)
    return XXs,yys
```

### 预测函数

```python
def splitPredictsOfBatch(predicts,numOfCellsInOneOfBatch):
    """
    function:
        对一个batch的预测结果进行拆分
    parameters:
        predicts：batch对应的预测结果
        numOfCellsInOneOfBatch：batch中每个拼接图像中所包含的细胞个数
    """
    results = []
    predicts = list(predicts)
    for i in range(len(predicts)):  # 依次拆分每一个拼接图片的预测结果
        predict_ = predicts[i]  # 某个拼接图片的预测结果
        num_ = numOfCellsInOneOfBatch[i]  # 某拼接图片中包含的细胞个数
        for j in range(num_):
            start = j*cfg0.class_num
            end = start+cfg0.class_num
            assert end<=len(predict_), "函数splitPredictsOfBatch：num_超过了拼接图片所能包含的细胞图片数量上限！"
            results_tmp = predict_[start:end]
            results.append(results_tmp)
    return results

def predicts_splice(sess,sess_op,inputs_placeholder,is_training_placeholder,X_input,batch_size,dataenhance,reDir):
        """
       functions:
           调用tf模型对X_input进行预测，返回预测结果，并统计预测所消耗的时长，具体通过如下步骤进行预测：
               1、首先按照样本名称将待预测图片拆分到不同的列表中
               2、依次对所有样本进行如下处理：
                   a、将某个样本的待预测图片按照cfg0.image_row*cfg0.image_column进行拼接
                   b、将拼接结果输入到模型中得到预测结果
                   c、将预测结果进行拆分得到每一个待预测样本对应的预测结果（列表长度等于参与拼接的待预测样本的数量）
               3、将所有的预测结果保存在同一个列表表中作为返回结果
       patameters:
           sess:当前模型所在的Session
           sess_op:待计算的模型最后一层的预测结果
           inputs_placeholder:待预测样本的placeholder
           is_training_placeholder:表示训练状态的placeholder
           X_input:待预测样本的路径集合
           batch_size:对X_input进行逐batch预测所采用的batch_size大小
           dataenhance:DataEnhance实例，用于读取样本
           reDir:输出信息文件
       returns:
           X_input所对应的预测结果列表、预测所有样本所消耗的总时间
        """
        time_start = time.time()  # 起始时间
        steps = 0  # 单个batch训练次数计数
        time_total = 0
        read_time = 0
        result = []
        batchRegister = []  # 用于存储一个batch
        numOfCellsInOneOfBatch = []  # batch中每个拼接图片所包含的细胞图片个数（用于拆分预测结果）
        sampleNames,samplesSplitByName = splitSampleByName(X_input)  # 输入列表按照样本划分
        
        # 首先生成样本标签
        result_label = []
        samplesSplitByName_tmp = []
        for i in range(len(samplesSplitByName)):
            samplesSplitByName_tmp = samplesSplitByName_tmp + samplesSplitByName[i]
        result_label = utils_.geneLabel(samplesSplitByName_tmp,labels)
        
        for i in range(len(samplesSplitByName)):  # 依次对所有样本进行处理
            cellOfSample = samplesSplitByName[i]  
            for j in range(0,len(cellOfSample),cfg0.image_row*cfg0.image_column):
                start = j
                end = min(j + cfg0.image_row*cfg0.image_column,len(cellOfSample))
                img_tmp,read_time_tmp = splice(cellOfSample[start:end],is_training=False)  # 进行图像读取、拼接
                read_time = read_time + read_time_tmp
                batchRegister.append(img_tmp)
                numOfCellsInOneOfBatch.append(end-start)
                if len(batchRegister)==batch_size or (i==len(samplesSplitByName)-1 and end==len(cellOfSample)):  # batchRegister中元素足够一个batch，进行预测
                    if steps%100==0:
                        utils_.printRd("myPredicts:"+str(steps),reDir)
                    # 进行预测 
                    result_tmp = sess.run(sess_op,feed_dict={inputs_placeholder:batchRegister,is_training_placeholder:False}) 
                    steps = steps + 1  # 计数器加一  
                    
                    # 对预测结果进行拆分，并保存到结果列表result中
                    result_tmp_split = splitPredictsOfBatch(result_tmp,numOfCellsInOneOfBatch)
                    result_tmp_split = list(result_tmp_split)
                    result = result+result_tmp_split 
                    # 清空batch及batch内部细胞计数列表
                    batchRegister = []
                    numOfCellsInOneOfBatch = []
                else:
                    pass
        time_end = time.time()  # 结束时间
        time_total = time_end-time_start
        return np.array(result),result_label,time_total-read_time
        
```

### 模型评价函数

```python
# 分析混淆矩阵
def analysisConfusionMatrix(ConfusionMatrix):
    result = ConfusionMatrix
    # 垃圾污染阳性
    two = float(sum(result[0][2:3]))/float(sum(result[0][:3])+1e-9)
    # 阴性污染阳性
    three = float(sum(result[1][2:3]))/float(sum(result[1][:3])+1e-9)
    # 阳性丢失
    one = float(sum(result[2][:2]))/float(sum(result[2][:3])+1e-9)
    print "阳性丢失："+str(round(one,5))+"  垃圾污染阳性："+str(round(two,5))+"  阴性污染阳性："+str(round(three,5))
    return one,two,three
```

### 定义F_beta_score

```python
# 分析：对于“异常类别”，和查准率相比较，查全率更加重要（查全率高才能降低假阴性率），所以beta应该大于一。 
#       但是也不能一味地追求查全率，查准率也有一定的重要性，所以使用使用“异常类别”的f_beta_score系数作为模型评价指标。
def F_beta_score(confusionMatrix,beta):  # confusionMatrix二分类任务混淆矩阵,beta为异常类别f1score系数。
    tp = confusionMatrix[0][0]
    fn = confusionMatrix[0][1]
    fp = confusionMatrix[1][0]
    tn = confusionMatrix[1][1]
    
    # 计算“正常类别”的F_beta值
    # p_normal = tp/(tp+fp)
    # r_normal = tp/(tp+fn)
    # f_beta1 = (1+beta1*beta1)*p_normal*r_normal/((beta1*beta1*p_normal)+r_normal)
    
    # 计算“异常类别”的F_beta值
    p_ab = tn/(tn+fn+0.00000001)
    r_ab = tn/(tn+fp+0.00000001)
    f_beta2 = (1+beta*beta)*p_ab*r_ab/((beta*beta*p_ab)+r_ab+0.00000001)
    
    return f_beta2

# 合并三分类混淆矩阵中“正常类别”和“垃圾类别”，然后调用上述F_beta_score
def F_beta_score_MergeNormalAndJunk(confusionMatrix,beta):
    newConfusionMatrix = []
    tp = confusionMatrix[0][0]+confusionMatrix[0][1]+confusionMatrix[1][0]+confusionMatrix[1][1]
    fn = confusionMatrix[0][2]+confusionMatrix[1][2]
    fp = confusionMatrix[2][0]+confusionMatrix[2][1]
    tn = confusionMatrix[2][2]
    
    newConfusionMatrix_one = []
    newConfusionMatrix_one.append(tp)
    newConfusionMatrix_one.append(fn)
    
    newConfusionMatrix_two = []
    newConfusionMatrix_two.append(fp)
    newConfusionMatrix_two.append(tn)
    
    newConfusionMatrix.append(newConfusionMatrix_one)
    newConfusionMatrix.append(newConfusionMatrix_two)
    
    return F_beta_score(newConfusionMatrix,beta)

import numpy as np
def F_beta_score_MergeNormalAndJunk2(confusionMatrix,beta=2,alpha=4):
    """
    confusionMatrix：混淆矩阵
    beta：表示阳性类别查全率是查准率的多少（beta*beta）倍。
    alpha：计算阳性类别查准率，Junk误分为阳性的惩罚是Negative误分为阳性惩罚的多少倍。
    备注：默认参数表示Positive类别的查全率是查准率重要性的4倍，Junk类别阳性污染率惩罚是Negative类别阳性污染率的4倍。
    """
    a = np.array(confusionMatrix)
    # 计算阳性类别查全率
    recall = a[2][2]/(sum(a[2])+1e-9)
    # 计算阳性类别的查准率 
    jp = a[0][2]/(sum(a[:,2])+1e-9)  # Junk的原始阳性污染率 
    np_ = a[1][2]/(sum(a[:,2])+1e-9)  # Negative的原始阳性污染率       
    precision = 1 - alpha*jp - np_  # 表示一个Negative误分了就算一个误分，但是一个Junk误分算alpha个误分
    f_beta = (1+beta*beta)*precision*recall/((beta*beta*precision)+recall+1e-9)
    return f_beta
```

```python
import time
import Build
import utils
import tensorflow as tf  
from IPython.display import display_html

def trainFun(utils,X_train,y_train,X_train_evalua,y_train_evalua,dataenhance,reDir='info/info.txt'):
    time_total = 0
    utils.printRd("",reDir,"w+")  # 清空重定向文件 
    tf.reset_default_graph()    # 清空运行图
    batch_size = cfg0.batch_size   
     # 实例化模型
    model = Build.SpliceNet(class_num=cfg0.class_num,dropout_rate=cfg0.dropout_rate)    
    
    # 构造正则化损失函数
    cost = model.cost  # 基本的交叉熵损失函数
    l2_loss = tf.add_n([tf.nn.l2_loss(var) for var in tf.trainable_variables()])   # L2正则化  
    if cfg0.weight_decay_rate!=None:
        cost = cost + l2_loss*cfg0.weight_decay_rate 

    # 表示训练多少个batch对应一epoch 
    epoch = int(len(X_train)/batch_size+1e-9)  
    epoch = int(epoch/(cfg0.image_row*cfg0.image_column))  # 现在一个epoch相当于以前的100个epoch，所以epoch应该变为原始epoch的一百分之一。
    # 设置训练轮数
    STEPS = int(epoch*cfg0.epochs)   
    print "the number of total batch:"+str(STEPS)
    
    # 配置tensorflow实用显存的方式
    config = tf.ConfigProto(allow_soft_placement=True)  # 自动选择可以运行的设备 
    config.gpu_options.allow_growth = True  # Gpu内存按需增长 
    # config.gpu_options.per_process_gpu_memory_fraction = 0.3 
    saver = tf.train.Saver(max_to_keep=1) # 模型保存器 
    
    # 使用Adam优化器，并且使用权重衰减
    global_step = tf.Variable(0)
    lr_decay_step = int(STEPS/(cfg0.lr_decay_times+1e-9))  # 注意：在拼接训练模型中此处无需修改
    learning_rate = tf.train.exponential_decay(cfg0.lr,global_step,lr_decay_step,cfg0.lr_decay_rate,staircase=True)
    # train = tf.train.GradientDescentOptimizer(learning_rate).minimize(model.cost,global_step=global_step)
    train = tf.train.AdamOptimizer(learning_rate).minimize(model.cost,global_step=global_step) 
    # train = tf.train.MomentumOptimizer(learning_rate,momentum=0.9,use_nesterov=True).minimize(model.cost,global_step=global_step) 

    with tf.Session(config = config) as sess:    
        sess.run(tf.global_variables_initializer())  # 初始化变量  
        
        # 判断是否加载与训练模型
        if cfg0.loadmodel_path!=None:
            saver = tf.train.import_meta_graph(cfg0.loadmodel_path+cfg0.loadmodel_name)
            saver.restore(sess,tf.train.latest_checkpoint(cfg0.loadmodel_path))

        val_accuracy_list = []  # 绘制验证集准确率变化曲线
        val_f1_list = []  # 绘制验证集f1_score变化曲线
        train_accuracy_list = []  # 绘制训练集准确率变化曲线
        cost_list_train = []
        cost_list_val = []
        cost_change = []  # 优化器所优化的每个batch的损失函数变化量
        cost_ = []  # 优化器所优化的每个batch的损失函数变化量

        val_f_beta_score_max = -1 # 记录val_f1的最大值
        val_f1_max = -1 # 记录val_f1的最大值
        for i in range(0,STEPS):   
            utils.printRd("............................第"+str(i)+"个batch............................",reDir)  
            if i%(int(cfg0.saveStepEpochRate*epoch))==0 and (i!=0 or cfg0.firstPred==True):  # 判断在进行训练之前是否进行预测。 
                
                
                # 处理验证集
                val_predicts,y_val,time_result = predicts_splice(sess,model.prob,model.inputs,model.is_training,X_val,batch_size,dataenhance,reDir)
                
                val_realLabel = utils.onehot2realLabel(y_val)
                val_predicts_realLabel = utils.onehot2realLabel(val_predicts,utils.countSample(X_train,labels))
                val_confusion_matrix = confusion_matrix(val_realLabel,val_predicts_realLabel) 
                print "测试集混淆矩阵为："
                print np.array(test_confusion_matrix)
                return
                
                
                
                # 记录“训练集准确率”、“验证集准确率”、“验证集F1”  
                train_predicts,y_train_evalua,_ = predicts_splice(sess,model.prob,model.inputs,model.is_training,X_train_evalua,batch_size,dataenhance,reDir)  
                val_predicts,y_val,_ = predicts_splice(sess,model.prob,model.inputs,model.is_training,X_val,batch_size,dataenhance,reDir)
                
                # 求真实标签
                train_evalua_realLabel = utils.onehot2realLabel(y_train_evalua)
                train_predicts_realLabel = utils.onehot2realLabel(train_predicts,utils.countSample(X_train,labels)) 
                
                val_realLabel = utils.onehot2realLabel(y_val)
                val_predicts_realLabel = utils.onehot2realLabel(val_predicts,utils.countSample(X_train,labels))

                # 计算“准确率”等 
                train_accuracy = accuracy_score(train_evalua_realLabel,train_predicts_realLabel, normalize=True)
                val_accuracy = accuracy_score(val_realLabel,val_predicts_realLabel, normalize=True)
                val_f1 = f1_score(val_realLabel,val_predicts_realLabel,average="macro") 
                train_confusion_matrix = confusion_matrix(train_evalua_realLabel,train_predicts_realLabel)
                val_confusion_matrix = confusion_matrix(val_realLabel,val_predicts_realLabel)
                
                # 修改模型评价指标为“异常类别”的f_beta_score
                val_f_beta_score = F_beta_score_MergeNormalAndJunk2(val_confusion_matrix,beta=2,alpha=4)  # beta=2，表示“异常类别”查全率的重要性是查准率的4倍
                
                # 训练集交叉熵
                # train_predicts_ = tf.nn.softmax(train_predicts)  # 交叉熵是基于softmax输出概率值的基础上的
                # val_predicts_ = tf.nn.softmax(val_predicts)
                cross_entropy_train = sess.run(-tf.reduce_mean(y_train_evalua*tf.log(tf.clip_by_value(train_predicts,1e-10,1.0))))
                cross_entropy_val = sess.run(-tf.reduce_mean(y_val*tf.log(tf.clip_by_value(val_predicts,1e-10,1.0))))
                
                # 对各个参数只保留五位小数（后面的位四舍五入处理）
                train_accuracy = round(train_accuracy,5)
                val_accuracy = round(val_accuracy,5)
                cross_entropy_train = round(cross_entropy_train,5)
                cross_entropy_val = round(cross_entropy_val,5)
                val_f1 = round(val_f1,5) 
                val_f_beta_score = round(val_f_beta_score,5)

                utils.printRd("训练集混淆矩阵为：",reDir)
                utils.printRd(np.array(train_confusion_matrix),reDir)
                utils.printRd(analysisConfusionMatrix(train_confusion_matrix),reDir)
                utils.printRd("验证集混淆矩阵为：",reDir)
                utils.printRd(np.array(val_confusion_matrix),reDir)
                utils.printRd(analysisConfusionMatrix(val_confusion_matrix),reDir)

                # 计算模型的训练速度
                speed = 1000*time_total/(i+1e-9)  # 训练一个batch的平均时长
                speed = speed/batch_size/(cfg0.image_column*cfg0.image_row)  # 训练一个样本的平均时长
                # speed = round(speed,4)
                utils.printRd("模型训练一个样本的平均时长为："+str(speed)+"毫秒",reDir)
                utils.printRd("第"+str(i)+"次迭代："+"train_accuracy="+str(train_accuracy)+" val_accuracy="+str(val_accuracy)+" cost="+str(cross_entropy_train)+" val_f1:"+str(val_f1)+ " val_f_beta_score:"+str(val_f_beta_score),reDir) 
 
                train_accuracy_list.append(train_accuracy)
                val_accuracy_list.append(val_accuracy)
                val_f1_list.append(val_f1) 
                cost_list_train.append(cross_entropy_train)
                cost_list_val.append(cross_entropy_val)
                lists = [val_accuracy_list,val_f1_list,train_accuracy_list]
                names = ["val_accuracy","val_f1","train_accuracy"]
                utils.showPerformOrCostCurve(lists,names,os.path.dirname(reDir)+'/perform.jpg') # 绘制准确率/F1分数
                utils.writeCurveValue(lists,names,os.path.dirname(reDir)+'/perform.txt') # 将准确率/F1分数列表写入文件
                lists = [cost_list_train,cost_list_val]
                names = ["train_cost","val_cost"]
                utils.showPerformOrCostCurve(lists,names,os.path.dirname(reDir)+'/cost.jpg') # 绘制损失函数
                utils.writeCurveValue(lists,names,os.path.dirname(reDir)+'/cost.txt') # 将损失函数列表写入文件 
                
                if val_f1>val_f1_max:
                    val_f1_max = val_f1
                    utils.printRd("val_f1最大值更新："+str(val_f1_max),reDir)

                # 修改模型保存条件为：每次epoch判断一次val_f1是否大于val_f1_max,假如满足条件就保存一次模型
                if val_f_beta_score>val_f_beta_score_max:  # 这样做也就相当于earlystoping。
                    # vgg.save_npy(sess,pwd+'jsL/keras_model/vggPara-save_train_l2正则化解决过拟合.npy')     
                    # 因为模型中使用了BN层，参数无法提取，所以替换使用Saver.save函数保存模型
                    saver.save(sess, cfg0.saveModelPath)  # 保存模型  
                    val_f_beta_score_max = val_f_beta_score
                    utils.printRd("model saved!",reDir)
                else:
                    utils.printRd("model not saved!",reDir)  

            # 每次选取batch_size个样本进行训练
            # 梯度下降训练模型参数 
            time_x = time.time()
            xxx,yyy = getMiniBatch(samplesSplitByName,is_training=True)  # 获取一个Batch的样本 
            time_y = time.time()
            
            # yyy_predict = sess.run(model.prob,feed_dict={model.inputs: xxx, model.is_training: False})
            # cost_batch_trainBefore =  sess.run(-tf.reduce_mean(yyy*tf.log(tf.clip_by_value(yyy_predict,1e-10,1.0))))
            time_start = time.time()  # 起始时间
            cost_tmp,_ = sess.run([model.cost,train], feed_dict={model.inputs: xxx, model.labels: yyy, model.is_training: True})  
            time_end = time.time()  # 结束时间
            print ("时间对比："+str(time_y-time_x)+"   "+str(time_end-time_start))
            # yyy_predict = sess.run(model.prob,feed_dict={model.inputs: xxx, model.is_training: False})
            # cost_batch_trainAfter =  sess.run(-tf.reduce_mean(yyy*tf.log(tf.clip_by_value(yyy_predict,1e-10,1.0)))) 
            # cost_change.append(cost_batch_trainAfter-cost_batch_trainBefore)
            cost_.append(cost_tmp)
            
            # lists = [cost_change,cost_,[0]*len(cost_change)]
            # names = ["batch_change_cost","batch_cost","基准线"]
            lists = [cost_]
            names = ["batch_cost"]
            utils.showPerformOrCostCurve(lists,names,os.path.dirname(reDir)+'/batch_cost.jpg') # 绘制损失函数
            utils.writeCurveValue(lists,names,os.path.dirname(reDir)+'/batch_cost.txt') # 将损失函数列表写入文件 
            
            
            time_total = time_total + time_end - time_start  # 累加模型训练的时间 
        display_html("<script>Jupyter.notebook.kernel.restart()</script>",raw=True)
        sess.close() 
```
